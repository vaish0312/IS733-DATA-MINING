```python
## Here we will check the percentage of nan values present in each columns

## Step 1 - Make the list of features which has missing values

columns_with_na = [columns for columns in df.columns if df[columns].isnull().sum()>1]

## Step 2 - Print the column name and percentage of missing values

for column in columns_with_na:
    print(column, np.round(df[column].isnull().mean(), 4)*100, '% missing values')
```
